```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('/content/housing-4.csv')
categorical_columns = df.select_dtypes(include=['object']).columns
if len(categorical_columns) >= 2:
    plt.figure(figsize=(10, 5))
    df[categorical_columns[1]].value_counts().plot(kind='bar', title=f'Bar Chart of {categorical_columns[1]}')
    plt.xlabel(categorical_columns[1])
    plt.ylabel('Frequency')
    plt.grid(axis='y')
    plt.tight_layout()
    plt.savefig('categorical_column_1_bar_chart.png')
    plt.show()
    plt.figure(figsize=(8, 8))
    df[categorical_columns[2]].value_counts().plot(kind='pie', autopct='%1.1f%%', title=f'Pie Chart of {categorical_columns[2]}')
    plt.ylabel('')
    plt.tight_layout()
    plt.savefig('categorical_column_2_pie_chart.png')
    plt.show()
else:
    print("No categorical columns found in the dataset.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = '/content/housing-4.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(df.head())

#Drawing Graphs with Numeric Columns
numeric_columns = df.select_dtypes(include=['number']).columns

if len(numeric_columns) >= 2:
    plt.figure(figsize=(10, 5))
    df[numeric_columns[1]].plot.hist(bins=20, title=f'Histogram of {numeric_columns[1]}')
    plt.xlabel(numeric_columns[1])
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('numeric_column_histogram.png')
    plt.show()

    plt.figure(figsize=(10, 5))
    df.plot.scatter(x=numeric_columns[1], y=numeric_columns[2], title=f'Scatter Plot of {numeric_columns[1]} vs {numeric_columns[2]}')
    plt.xlabel(numeric_columns[1])
    plt.ylabel(numeric_columns[2])
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('numeric_column_scatter.png')
    plt.show()
else:
    print("No numeric columns found in the dataset.")

# Creating a categorical column for housing_median_age
bins = [0, 10, 20, 30, 40, 50, 60]
labels = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-60']
df['age_group'] = pd.cut(df['housing_median_age'], bins=bins, labels=labels, right=False)

# Drawing Graphs with Categorical Columns
categorical_columns = df.select_dtypes(include=['category', 'object']).columns

if 'age_group' in df.columns:
    plt.figure(figsize=(10, 5))
    df['age_group'].value_counts().sort_index().plot(kind='bar', title='Bar Chart of Housing Median Age Groups')
    plt.xlabel('Age Group')
    plt.ylabel('Count')
    plt.xticks(rotation=0)
    plt.grid(axis='y')
    plt.tight_layout()
    plt.savefig('categorical_column_bar_chart.png')
    plt.show()

    plt.figure(figsize=(8, 8))
    df['age_group'].value_counts().plot(kind='pie', startangle=140, title='Pie Chart of Housing Median Age Groups')
    plt.ylabel('')
    plt.tight_layout()
    plt.savefig('categorical_column_pie_chart.png')
    plt.show()
else:
    print("No categorical columns found in the dataset.")
```

```python
import tensorflow as tf
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
import datetime

# Load the 2D MNIST images from TensorFlow
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Flatten each 2D MNIST image into a 1D array
x_train_flatten = x_train.reshape(x_train.shape[0], -1)
x_test_flatten = x_test.reshape(x_test.shape[0], -1)

# Convert multi-class labels into binary class for training and test sets
y_train_5 = (y_train == 5)
y_test_5 = (y_test == 5)

# Feature scaling using standardization (z-score normalization)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train_flatten)
x_test_scaled = scaler.transform(x_test_flatten)

# Print the current date and time
now = datetime.datetime.now()
print("Running the lab04 step6 at date and time: ")
print(now.strftime("%Y-%m-%d %H:%M:%S"))

# Train the KNN classifier and report classification accuracy
KNN_classifier = KNeighborsClassifier(n_neighbors=3)
KNN_classifier.fit(x_train_scaled, y_train_5)

# Predict on the test set
y_test_pred = KNN_classifier.predict(x_test_scaled)

# Calculate and print accuracy
accuracy = metrics.accuracy_score(y_test_5, y_test_pred)
print("Accuracy with k=3 and feature scaling: ", accuracy)
```

```python
import tensorflow as tf
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.preprocessing import StandardScaler

# Load the 2D MNIST images from TensorFlow
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Flatten each 2D MNIST image into a 1D array
x_train_flatten = x_train.reshape(x_train.shape[0], -1)
x_test_flatten = x_test.reshape(x_test.shape[0], -1)

# Convert multi-class labels into binary class for training and test sets
y_train_5 = (y_train == 5)
y_test_5 = (y_test == 5)

# Feature scaling using standardization (z-score normalization)
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train_flatten)
x_test_scaled = scaler.transform(x_test_flatten)

# Try different K values and report their test accuracy
k_values = [1, 3, 5, 7, 9]
accuracies_with_scaling = {}

for k in k_values:
    KNN_classifier = KNeighborsClassifier(n_neighbors=k)
    KNN_classifier.fit(x_train_scaled, y_train_5)
    y_test_pred = KNN_classifier.predict(x_test_scaled)
    accuracy = metrics.accuracy_score(y_test_5, y_test_pred)
    accuracies_with_scaling[k] = accuracy
    print(f"Accuracy with k={k} and feature scaling: ", accuracy)

# Reporting the accuracies for different k values with feature scaling
print("Accuracies for different k values with feature scaling: ", accuracies_with_scaling)

```

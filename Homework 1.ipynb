```python
#task 1
#simulate the input and output datasets
np.random.seed(42)

#generate random values for N and D
#random integers between 500 and 1000
N = np.random.randint(500,1001)
#random integers between 10 and 100
D = np.random.randint(10,101)

#generate the input dataset (N*D)
x = np.random.rand(N,D)

#generate the output dataset (N*1) between 1 and 10
y = np.random.uniform(1,10,size=(N,1))

#task 2
#split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2, random_state =42)

#initialising and training the regression model
model = LinearRegression()
model.fit(x_train,y_train)

#predictig the test set
y_pred = model.predict(x_test)

#calculating the metrics
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

#output
print("Dataset dimensions:")
print(f"Input dataset (x): {x.shape}")
print(f"Output dataset (y): {y.shape}")
print("\nEvaluation metrics:")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"R-squared (R2): {r2:.4f}")
```
